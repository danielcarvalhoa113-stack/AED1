```python
import random
n = random.randint(1, 50)
while True:
    a = int(input("Digite a:"))
    if a == n:
        print("Acertou! Parabéns!!")
        break
    else:
        print("Errou!")
        if a > n:
            print("Tente um número menor")
        else:
            print("Tente um número maior")
```

```python
import random
n_secreto = random.randint(1, 50)
for i in range(5):
    palpite = int(input("Digite um palpite: "))
    if palpite == n_secreto:
        print("Acertou!!")
        break
    else:
        if palpite < (n_secreto + 5) or palpite > (n_secreto - 5):
            print("Está quente")
        else:
            print("Está frio")
print("Acabou")
```
