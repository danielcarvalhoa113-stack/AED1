```python
n = int(input("digite a qtd de pessoas: "))
soma_peso = 0
soma_altura = 0
imc_maior = 0
imc_menor = 1000
x = 1
while x <= n:
    peso = float(input("Digite o peso, em kg:"))
    altura = float(input("Digite a altura, em metros:"))
    imc = peso/(altura ** 2)
    if imc < imc_menor:
        imc_menor = imc
    if imc > imc_maior:
        imc = imc
    soma_peso += peso
    soma_altura += altura
    x += 1
media_peso = soma_peso / n
media_altura = soma_altura / n
print("Imc maior =", imc_maior, "Imc menor =", imc_menor)
print("Peso médio =", media_peso)
print("Altura média =", media_altura)
```
